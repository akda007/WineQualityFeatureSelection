```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
plt.style.use("default")

df = pd.read_csv("WineQT.csv") 
df.head()

```

```python

print("\nVariaveis")
print(df.dtypes)

print("\nValores ausentes")
print(df.isna().sum())
```

```python

duplicadas = df.duplicated().sum()
print("Duplicatas encontradas:", duplicadas)

df = df.drop_duplicates()
print("Após remover duplicatas:", df.shape)
```

```python
if 'Id' in df.columns:
    print("\nMovendo coluna 'Id' para a primeira posição...")
 
    coluna_id = df.pop('Id')
   
    df.insert(0, 'Id', coluna_id)
else:
    print("\nColuna 'Id' não encontrada.")
```

```python

print("\nValores mínimos:")
print(df.min())

print("\nValores máximos:")
print(df.max())
```

```python
print("\nEstatísticas descritivas:")
print(df.describe().T)
```

```python
df.hist(figsize=(14,10), bins=20)
plt.suptitle("Distribuição das variáveis")
plt.show()
```

```python

df.to_csv("wine_clean.csv", index=False, float_format='%.2f')

print("Arquivo salvo: wine_clean.csv")
```
